## Step 1. Getting the data

Let's make a sample request to retrieve some questions posted on Stack Exchange on the first day of 2015. Documentation of the Stack Exchange API can be found [here](https://api.stackexchange.com/docs). You might need to install the *requests* library and an easy way to do so is by using the command [pip](https://docs.python.org/3/installing/index.html).

````python
import requests
from datetime import datetime

start_time = 1420070400 # 01-01-2015 at 00:00:00
end_time   = 1420156800 # 01-02-2015 at 00:00:00

response = requests.get("https://api.stackexchange.com/2.2/questions?pagesize=100" +
                        "&fromdate=" + str(start_time) + "&todate=" + str(end_time) +
                        "&order=asc&sort=creation&site=stackoverflow")
print(response)
type(response)
````

Now, object *response* contains the raw text of posts from the first day of 2015.

We can try to print the response that Stack Exchange returns.

````python
print(response.text)
````

## Step 2. Parsing the responses

In this section, we practice some of the basic Python tools that we learned in class and the powerful string handling methods that Python offers. Our goal is to be able to pick the interesting parts of the response and transform them in a format that will be useful to us.

First let's isolate the creation_date in the response. Fill in the rest of the ```print_creation_dates_json()``` function that reads the response and prints the creation dates and whether the question was answered. Your solution should be in the form **(creation_date,is_answered)**. Notice that a JSON object is basically a dictionary. **(2 pts)**

````python
def print_creation_dates_is_answered_json(response):
    filtered = response.json()
    for index in range(len(filtered["items"])): #for each item
        date = datetime.fromtimestamp(filtered["items"][index]["creation_date"]+ 18000).strftime('%c')#printing time 
        print("(" , date, "," , filtered["items"][index]["is_answered"] , ")", sep="") #print is_answered
    
````

Some hints/library you might find useful are the following:

````python
# File imports
import xml.etree.ElementTree as ET
import pandas as pd
import timeit

parsed = []
start = timeit.default_timer()
for event, elem in ET.iterparse("stackoverflow-posts-2015.xml"):
    if ('PostTypeId' not in elem.attrib):
        continue
    if (elem.attrib['PostTypeId'] == '1' or elem.attrib['PostTypeId'] == '2'): #check if its the appropriate post type
        if (elem.attrib['PostTypeId'] == '1'):
            PostType = 'Q'
        else:
            PostType = 'A'
            
            #gather rest of columns, adding none if its not present
        if ('Id' not in elem.attrib):
            Id = 'None'
        else:
            Id = elem.attrib['Id']

        if ('CreationDate' not in elem.attrib):
            CreationDate = 'None'
        else:
            CreationDate = elem.attrib['CreationDate']

        if ('OwnerUserId' not in elem.attrib):
            OwnerUserID = 'None'
        else:
            OwnerUserID = elem.attrib['OwnerUserId']

        if ('Score' not in elem.attrib):
            Score = 'None'
        else:
            Score = int(elem.attrib['Score'])

        if ('ViewCount' not in elem.attrib):
            ViewCount = 'None'
        else:
            ViewCount = elem.attrib['ViewCount']

        if ('AnswerCount' not in elem.attrib):
            AnswerCount = 'None'
        else:
            AnswerCount = elem.attrib['AnswerCount']

        if ('ParentId' not in elem.attrib):
            ParentID = 'None'
        else:
            ParentID = elem.attrib['ParentId']

        if ('Tags' not in elem.attrib):
            Tags = 'None'
        else:#string manipulation to get first tag
            preTag = elem.attrib['Tags']
            temp1 = preTag.replace("<", "")
            temp2 = temp1.split(">")
            Tags = str(temp2[0])

        parsed.append([Id, CreationDate, OwnerUserID, Tags, Score, ViewCount, AnswerCount, ParentID, PostType])

    elem.clear()
print (timeit.default_timer() - start)#timer

with open("parsed.txt", 'w') as writer:#writing to file
    for item in parsed:
        writer.write("{}\n".format(item))
````

````python
col = ["Id", "CreationDate", "OwnerUserId", "Tags", "Score", "ViewCount", "AnswerCount", "ParentId", "PostType"]
df = pd.DataFrame(parsed, columns=col)
pd.set_option('display.expand_frame_repr', False)#creating df
#d.set_option('display.max_rows', 5740132)
print(df)
````

## Step 3. Putting it all together

We are now ready to tackle our original problem. Write a function to measure the similarity of the top 100 users with the highest total answer scores. That is, you need to compute for each user the sum of the answer scores that he has received, for all answers that he has provided.

Compare the users based on the types of questions they answer. We will categorize a question by its first tag, so eventually the features of the user will comprise the first tag of each question that he has answered. You may choose to implement any one of the similarity/distance measures we discussed in class. **(2pts)**

Note that answers are posts with `PostTypeId=2`. The ID of the question in answer posts is the `ParentId`.

You may find the [sklearn.feature_extraction](http://scikit-learn.org/stable/modules/feature_extraction.html) module helpful.

````python
qs = df[df.PostType == 'A']#get all Answers
grouped = qs.groupby("OwnerUserId")#group the same users
summed = grouped.sum()#sum the score
ordered = summed.sort_values(by='Score', ascending=False)[:100]#desceding order, only top 100
print(ordered)
````

````python
taglist = []
ordDict = ordered.index.values

dfDict = df[['Id', 'Tags']].set_index('Id')
dfDict = dfDict.to_dict('dict')

for index in range(len(ordDict)): #for each of top 100
    UIDstring = ordDict[index] #gets the OwnerUserId from string manipulation
    
    byUser = (qs.loc[qs['OwnerUserId'] == UIDstring])['ParentId'] #set of parents from set of answers
    byUserList = byUser.tolist()
    
    tempstr = ""
    tags = []
    for index2 in byUserList:
        try:
            tag = dfDict['Tags'][index2] #sometimes parent not in the set of data
            tags.append(str(tag))
        except:
            pass
        
    tempstr = " ".join(tags)
    taglist.append(tempstr)#gets the tags of a user  
print(taglist)
print(len(taglist))
````

````python
#convert to vectors
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(analyzer='char_wb', ngram_range=(1,200))#vectorizer is in char_wb mode to preserve punctuation (c++, etc)

X = vectorizer.fit_transform(taglist)
y = vectorizer.get_feature_names()
print((y))

tagarray = X.toarray()
print(tagarray)
````

````python
#cosine similarity
from sklearn.metrics.pairwise import cosine_similarity

#gets cosine similarity of each user compared to all 100 users
data = []
for index in range(len(tagarray)):
    cosinevals = []
    for index2 in range(len(tagarray)):
        set1 = tagarray[index].tolist()
        set2 = tagarray[index2].tolist()
        temp = ((cosine_similarity([set1], [set2])).tolist())
        cosinevals.append(temp[0][0])
    data.append(cosinevals)

print((data))
````

Plot the distance of the top 100 users using a [heatmap](https://stanford.edu/~mwaskom/software/seaborn/generated/seaborn.heatmap.html). **(1 pts)**
